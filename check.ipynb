## Coordinate Matching

```python
#latlong matching:
latdelta=[]
longdelta=[]
for lat1,long1,lat2,long2 in zip(osm_match['latitude'],osm_match['longitude'],google_match['latitude'],google_match['longitude']):
    latdelta.append(abs(lat1-lat2))
    longdelta.append(abs(long1-long2))
```

```python
for idx,_ in enumerate(longdelta):
    print(latdelta[idx], end=', ');print(longdelta[idx])
```

```python
#find delta distance:
import math
distancedelta=[]
for idx,_ in enumerate(longdelta):
    distancedelta.append(math.sqrt(latdelta[idx]**2 + longdelta[idx]**2))

distancedelta
```

## Name Matching

```python
# python builtin:
from difflib import SequenceMatcher
score_sequencer=[]
for osmname,googname in zip(osm_match['name'],google_match['name']): 
    googname=str(googname)
    osmname=str(osmname)

    print("comparing "+osmname+ " and "+googname)
    s = SequenceMatcher(None, osmname, googname)
    print(s.ratio())
    score_sequencer.append(s.ratio())
```

## Category Matching

```python
category_match=[]
from difflib import SequenceMatcher

for osmname,googname in zip(osm_match['categories'],google_match['categories']): 
    googname=str(googname)
    osmname=str(osmname)
    print("comparing "+osmname+ " and "+googname)
    s = SequenceMatcher(None, osmname, googname)
    print(s.ratio())
    category_match.append(s.ratio())
```
