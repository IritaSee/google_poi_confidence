## Preparations

```python
for idx,a in enumerate(matching['osm_id']):
    if(idx==0):osm_match = pd.DataFrame(osm.query(" osm_id == "+str(a)))
    else:osm_match = pd.concat([osm_match,pd.DataFrame(osm.query(" osm_id == "+str(a)))])
```

```python
pd.options.display.max_columns = None
osm_match
```

```python
for idx,a in enumerate(matching['internal_id']):
    if(idx==0):google_match = pd.DataFrame(google.query("internal_id == "+"'"+str(a)+"'"))
    else:google_match = pd.concat([google_match,pd.DataFrame(google.query("internal_id == "+"'"+str(a)+"'"))])
```

```python
pd.options.display.max_columns = None
google_match
```

## Coordinate Matching

```python
#latlong matching:
latdelta=[]
longdelta=[]
for lat1,long1,lat2,long2 in zip(osm_match['latitude'],osm_match['longitude'],google_match['latitude'],google_match['longitude']):
    latdelta.append(abs(lat1-lat2))
    longdelta.append(abs(long1-long2))
```

```python
for idx,_ in enumerate(longdelta):
    print(latdelta[idx], end=', ');print(longdelta[idx])
```

```python
#find delta distance:
import math
distancedelta=[]
for idx,_ in enumerate(longdelta):
    distancedelta.append(math.sqrt(latdelta[idx]**2 + longdelta[idx]**2))
#the greater the worse
distancedelta
```

## Name Matching

### Built in Sequencer

```python
# python builtin:
from difflib import SequenceMatcher
score_sequencer=[]
for osmname,googname in zip(osm_match['name'],google_match['name']): 
    googname=str(googname).lower()
    osmname=str(osmname).lower()

    print("comparing "+osmname+ " and "+googname)
    s = SequenceMatcher(None, osmname, googname)
    print(s.ratio())
    score_sequencer.append(s.ratio())
```

```python
#using sequence matcher seems fast, but when we need to pick a treshold, the result seems not quite accurate, 
# lets try Levenshtein Distance [https://en.wikipedia.org/wiki/Levenshtein_distance]

#we can find from previous result that some places name with extra words (ex: nuno vs nuno pizerria) decrease the ratio, so lets find them:
from thefuzz import fuzz
fuzz_name=[]
is_with_subs=[]
for osmname,googname in zip(osm_match['name'],google_match['name']): 
    googname=str(googname).lower()
    osmname=str(osmname).lower()
    
    print("comparing "+osmname+ " and "+googname)
    test=fuzz.token_sort_ratio(osmname,googname)
    print(test/100)
    fuzz_name.append(test/100)
    if ((osmname.find(googname) !=-1 or googname.find(osmname) != -1) and test!=100):
        is_with_subs.append(True)
    else : is_with_subs.append(False)

```

## Category Matching

```python
from thefuzz import fuzz
fuzz_cat=[]
for osmname,googname in zip(osm_match['categories'],google_match['categories']): 
    googname=str(googname).lower()
    osmname=str(osmname).lower()
    googname=googname[1:-1]
    osmname=osmname[1:-1]
    if(googname == 'a'):googname='---'
    if(osmname == 'a'):osmname='---'
    print("comparing "+osmname+ " and "+googname)
    test=fuzz.token_sort_ratio(osmname,googname)
    #just like name matching but simpler, and weighted lower than name matching, (for me, just add ten points)
    if ((osmname.find(googname) !=-1 or googname.find(osmname) != -1) and test!=100):
       test=test+10

    print(test/100)
    fuzz_cat.append(test/100)
```

## Address Matching

```python
#we know that both have different format, but for the sake of time, we are going to just simply join all address related from OSM, and compare 
#it with google's

from thefuzz import fuzz

#address related in OSM: 
addr_type = [
'address_house_nr', 'address_street',	'address_zip_code',	'address_city',	
'address_country',	'address_full',	'address_region_neighborhood',	'address_region_suburb',	
'address_region_district',	'address_region_province',	'address_region_state',	'address_house_name',	
'address_place',	'address_block',	'address_details_level',	'address_details_flats',	'address_details_unit'
]
# let's join them!
joined_addr=[]
for index,_ in enumerate(osm_match['osm_id']):
    temp=""
    for types in addr_type:
        # if(osm_match[types].iloc[index] != math.nan ):temp=temp+str(osm_match[types].iloc[index])+", "
        #temp=temp+str(osm_match[types].iloc[index])+", "
        if(str(osm_match[types].iloc[index]) != "nan" ):temp=temp+str(osm_match[types].iloc[index])+", "
        else: temp=temp+""
    temp=temp[:-2]
    #temp.replace('nan','')
    joined_addr.append(temp)

fuzz_addr=[]
addr_with_subs=[]
for osmname,googname in zip(joined_addr,google_match['address']): 
    googname=str(googname).lower()
    osmname=str(osmname).lower()
    googname=googname[1:-1]

    print("comparing "+osmname+ " AND "+googname)
    test=fuzz.token_sort_ratio(osmname,googname)
    print(test/100)
    fuzz_addr.append(test/100)
    if ((osmname.find(googname) !=-1 or googname.find(osmname) != -1) and test!=100):
        addr_with_subs.append(True)
    else : addr_with_subs.append(False)
## we can see the result is not quite good, that's the best for now :(
    ## we can improve the result by chopping the address but i am running out of time :( sorry
```

# Scoring 

```python
#create a dataframe to see everything in one overview

matching['latlong_score']=distancedelta
matching['name_score']=fuzz_name
matching['name_has_substring']=is_with_subs
matching['categories_score']=fuzz_cat
matching['address_score']=fuzz_addr
matching['addr_has_substring']=addr_with_subs
```

```python
#this requires more research, but since i am running out of time, here is what I got:
# 1) we need look further into the weighting system, A.K.A. How each scores represent the true final score, for now, I will just naively pick some score, in the code below
# 2) Address is the trickiest one, requires a lot of chopping, joining, etc
# 3) I could not access this useful research yet -> https://www.researchgate.net/publication/2834776_A_comparison_of_string_DISTANCE_metrics_for_name-matching_tasks

def delta_normalized(x):
    upper=max(distancedelta)
    lower=min(distancedelta)
    return (x-lower)/upper

latlong_weight=0.4
name_weight=0.3
is_with_subs_weight=0.1
category_weight=0.1
address_weight=0.05
addr_substring_weight=0.05

final=[]
for idx,_ in enumerate(matching['osm_type']):
    final.append(((1-delta_normalized(distancedelta[idx]))*latlong_weight) + (fuzz_name[idx])*name_weight + (int(is_with_subs[idx])*is_with_subs_weight) 
    + (fuzz_cat[idx]*category_weight) + (fuzz_addr[idx]*address_weight) + (addr_substring_weight*int(addr_with_subs[idx])))

final
```

```python
matching['confidence']=final
matching
```
